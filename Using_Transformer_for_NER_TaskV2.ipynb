```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
```

```python
checkpoints = "Davlan/xlm-roberta-base-ner-hrl"

tokenizer = AutoTokenizer.from_pretrained(checkpoints)
model = AutoModelForTokenClassification.from_pretrained(checkpoints)
```

```python
import pandas as pd

traffy_df = pd.read_csv("/content/drive/MyDrive/DSI444/name_csv_file.csv")
```

```python
traffy_df = traffy_df.rename(columns={"0": "comment"})
traffy_df
```

```python
import re

def replace_like_list(text_list):
  text_list = re.sub(r"\'", "", text_list)
  text_list = re.sub(r"\[", "", text_list)
  text_list = re.sub(r"\]", "", text_list)

  return text_list
```

```python
traffy_df["comment"] = traffy_df["comment"].apply(lambda x: replace_like_list(x))
traffy_df
```

```python
def clean_some_txt(text_list):
  text_list = re.sub(r"\,", "", text_list)
  text_list = re.sub(r"\(|\)", "", text_list)

  return text_list
```

```python
traffy_df["comment"] = traffy_df["comment"].apply(lambda x: clean_some_txt(x))
traffy_df
```

```python
import torch
import tqdm
```

```python
output_list = []

for comment in traffy_df["comment"]:
  inputs = tokenizer(comment, padding=True, truncation=True, return_tensors="pt")
  tokens = inputs.tokens()
  outputs = model(**inputs)
  predictions = torch.argmax(outputs.logits, dim=2)

  full_token = ""
  all_tag = []
  for token, prediction in zip(tokens, predictions[0].numpy()):
    full_token += token
    all_tag.append(model.config.id2label[prediction])
  
  output_list.append([full_token, all_tag])
```

```python
import pandas as pd

output_df = pd.DataFrame(data=output_list, columns=["word", "ner_tag"])
output_df
```
